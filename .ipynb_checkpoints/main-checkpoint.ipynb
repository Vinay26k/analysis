```python
#This part performs the stage 1 operations.

import pandas as pd
import numpy as np

df = pd.read_csv('data.csv')

def stage1(df):

    dummy = pd.DataFrame()

    a = ' C1) Funding FY2008 '
    b = ' C2) Funding FY2009 '

    pat = r"-"

    df[a] = df[a].str.replace(pat,'0')
    df[b] = df[b].str.replace(pat,'0')

    df= df.fillna('0')

    df['growth(%)'] = (df[b].str.replace(',','').astype('float') - df[a].str.replace(',','').astype('float'))*100/df[b].str.replace(',','').astype('float')


    df['target'] = (df['growth(%)'] > 0).astype('int') 
    
    return df

df = stage1(df)
print(df.head(15))
```

```python
import matplotlib.pyplot as plt
plt.rcParams.update({'figure.max_open_warning': 0})
import seaborn as sns
from sklearn.preprocessing import LabelEncoder

neg = [' C1) Funding FY2008 ',' C2) Funding FY2009 ',' C3) Funding FY2010 '] 
enc = LabelEncoder()

uniq = df.apply(pd.Series.nunique)
cols_to_drop = uniq[uniq == 1].index
df.drop(cols_to_drop,axis = 1,inplace = True)

def stage2_P1(df):
    return enc.fit_transform(df)

#univariate distribution of features.

for i,col in enumerate(df.columns):
    
    if col in neg:
        continue
    
    plt.figure(i)
    df[col] = df[col].astype('str')
    sns.distplot(stage2(df[col]),axlabel = col)
```

```python
from sklearn.metrics import mutual_info_score

y = df['target']
df.drop(['target'],axis = 1 , inplace = True)

def stage2_P2(df):
    return mutual_info_score(y,df)

for i,col in enumerate(df.columns):
    if col in neg:
        continue
    
    print("Mutual info score ("+str(col)+") = "+str(stage2_P2(df[col])))
```

```python
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import train_test_split

#S3P1 Splittting dataset into train and test samples.

X_train,X_test,y_train,y_test = train_test_split(df,y,test_size = 0.3,random_state = 123)

#S3P1 training model with xgboost.

from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score

df = df.apply(LabelEncoder().fit_transform)

model = XGBClassifier(objective = 'binary:logistic',seed = 23)
model.fit(X_train,y_train)
pred = model.predict(X_test)

y_test.reset_index(drop = True, inplace = True)
pred = np.array(pred.astype('int'))
y_test = np.array(y_test.astype('int'))

print(roc_auc_score(y_test,pred))
```
